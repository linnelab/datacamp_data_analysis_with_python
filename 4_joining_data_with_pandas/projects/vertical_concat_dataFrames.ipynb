# Vertical Concat DataFrames


## üîî Method:

üî∏ use **`concat()`**
    
* Can concatenate both **vertical** and **horizontal**, when **`axis=0`** is vertical.

* **Vertical concate mutiple tables**:
    * **`pd.concat([table_name1, table_name2, table_name3,..])`**
    * These tables can contain different column counts (i.e. part of columns name are same, some columns are diff). 
      If tables not contain this column, it will be auto marked **NaN** in rows.

* **Only columns in all tables**:
    * pd.concat([table_names], **`join='inner'`**)
    * The result table shows only columns names that are in all tables, 
      if some columns are only in some tables not all, this column will not be shown.

* **Ignore original index**:
    * pd.concat([table_names], **`ignore_index=True`**)
    * After concat multiple tables, it will return a result table, this table will maintain the original index, 
      can use this setting to ignore the original index, and rearrange the index of the result table.
    * the index goes from 0 to n-1.
      
* **Sort column names**:
    * pd.concat([table_names], **`sort=True`**)
    * ASC sorting for columns of cancat result table. 

* **Set table index keys**:
    * pd.concat([table_name], **`keys=['key_name1', 'key_name2', 'key_name3']`**)

* **Verifying data integrity** :
    * pd.concat(**`verify_integrity=True`**)
    * default value is **False**.
    * check whether the new concatenated **index** contains duplicates.
    * if contain duplicate index, it will show error message.
    * if **not** contain duplicate index, it will show concate table results.


üî∏ use **`append()`**

* Another method concate table simplified version.

* **Support**: 
    * **`ignore_index=`**
    * **`sort=`**
    
* ‚≠ê **Not Support**: 
    * **`join=`** 
    * **`keys=[]`**
    
* **Vertical concate mutiple tables**:
    * **<font color=red>table_name1</font>`.append([table_name2, table_name3,..])`**
    * These tables can contain different column counts (i.e. part of columns name are same, some columns are diff). 
      If tables not contain this column, it will be auto marked **NaN** in rows.

## üìì Read source data - csv file


* **Method**: use **`read_csv()`** to read csv source files.

* **Data Sources Link**: github datacamp_data_analysis_with_python repository:
  https://github.com/linnelab/datacamp_data_analysis_with_python/tree/main/4_joining_data_with_pandas/datasets/csv_files

* ‚≠ê **Github Repository Setting**: need to set **`public`** and use type= **`RAW`** link on github

```python
import pandas as pd

# source data link
link_tracks_master = 'https://raw.githubusercontent.com/linnelab/datacamp_data_analysis_with_python/main/4_joining_data_with_pandas/datasets/csv_files/tracks_master.csv'
link_tracks_ride = 'https://raw.githubusercontent.com/linnelab/datacamp_data_analysis_with_python/main/4_joining_data_with_pandas/datasets/csv_files/tracks_ride.csv'
link_tracks_st = 'https://raw.githubusercontent.com/linnelab/datacamp_data_analysis_with_python/main/4_joining_data_with_pandas/datasets/csv_files/tracks_st.csv'
link_inv_jul = 'https://raw.githubusercontent.com/linnelab/datacamp_data_analysis_with_python/main/4_joining_data_with_pandas/datasets/csv_files/inv_jul.csv'
link_inv_aug = 'https://raw.githubusercontent.com/linnelab/datacamp_data_analysis_with_python/main/4_joining_data_with_pandas/datasets/csv_files/inv_aug.csv'
link_inv_sep = 'https://raw.githubusercontent.com/linnelab/datacamp_data_analysis_with_python/main/4_joining_data_with_pandas/datasets/csv_files/inv_sep.csv'
link_inv_items = 'https://raw.githubusercontent.com/linnelab/datacamp_data_analysis_with_python/main/4_joining_data_with_pandas/datasets/csv_files/invoice_items.csv'

# read csv files
tracks_master = pd.read_csv(link_tracks_master, index_col=0)
tracks_ride = pd.read_csv(link_tracks_ride, index_col=0)
tracks_st = pd.read_csv(link_tracks_st, index_col=0)
inv_jul = pd.read_csv(link_inv_jul, index_col=0)
inv_aug = pd.read_csv(link_inv_aug, index_col=0)
inv_sep = pd.read_csv(link_inv_sep, index_col=0)
inv_items = pd.read_csv(link_inv_items, index_col=0)

# show table results
display(tracks_master)
display(tracks_ride)
display(tracks_st)

# show part of results
display(inv_jul.head())
display(inv_aug.head())
display(inv_sep.head())
display(inv_items.head())

# print table rows and columns count
print('tracks_master table =', tracks_master.shape)
print('tracks_ride table =', tracks_ride.shape)
print('tracks_st table =', tracks_st.shape)
print('inv_jul table =', inv_jul.shape)
print('inv_aug table =', inv_aug.shape)
print('inv_sep table =', inv_sep.shape)
print('inv_items table =', inv_items.shape)
```

## ‚úçÔ∏è Practice I - concat()
            
* Concatenate tables with different column counts.

```python
# vertical concate tables, rearrange index, and sort columns
tracks_from_albums = pd.concat([tracks_master, tracks_ride, tracks_st], ignore_index=True, sort=True)

# show results
display(tracks_from_albums)
```

```python
# showing only columns that are in all tables
tracks_from_albums = pd.concat([tracks_master, tracks_ride, tracks_st], join='inner', ignore_index=True, sort=True)

# show results
display(tracks_from_albums)
```

## ‚úçÔ∏è Practice II - concat()
            
* Would like to know which month in the quarter saw the highest average invoice total. 
  Concatenate these tables into one to create a graph of the average monthly invoice total.

```python
# concat three month tables and set index keys
invoice_three_month = pd.concat([inv_jul, inv_aug, inv_sep], keys=['7Jul', '8Aug', '9Sep'])

# show results
display(invoice_three_month)
```

```python
# Group the invoices by the index keys and find avg of the total column
avg_inv_by_month = invoice_three_month.groupby(level=0).agg({'total':'mean'})

# show results
display(avg_inv_by_month)
```

```python
import matplotlib.pyplot as plt

# Bar plot of avg invoices by month and rotation for ticks (xticks for vertical, yticks for horizontal bar plot)
avg_inv_by_month.plot(kind='bar', rot=0)
plt.show()
```

```python
# use verify_integrity=True to check whether the new concatenated index contains duplicates.
pd.concat([inv_jul, inv_aug], verify_integrity=True)
```

```python
# if set verify_integrity=False, it will not verify whether have duplicate index, then show concate result table.
pd.concat([inv_jul, inv_aug], verify_integrity=False)
```

## ‚úçÔ∏è Practice - append()
            
* Append these tables to determine which track sold the most quantity.

```python
# append tables and rearrange index
metallica_tracks = tracks_master.append([tracks_ride, tracks_st], ignore_index=True)

# show result
display(metallica_tracks)
```

* **Note**: 
    * If right table have two duplicate rows, then do inner join, result table only reserve one rows not show two rows.
      It will cause sum quantity happen incorrect values, 
      so I add iid this column and insert diff value on per rows to avoid this problem.

```python
# use append results table inner join invoice items table to get quantity
tracks_invoices = metallica_tracks.merge(inv_items, on='tid')

# show results
display(tracks_invoices)
```

```python
# group by tid and name, sum the quantity sold, and DESC sort quantity 
tracks_sold = tracks_invoices.groupby(['tid', 'name']).agg({'quantity':'sum'}).sort_values('quantity', ascending=False)

# show results
display(tracks_sold)
```
